### Парето-оптимизация

Прежде, чем проводить Парето-оптимизацию необходимо отмасштабировать данные в работоспобных вариантах, а также значения идеальных точек.

```r
# Функция нормировки датафрейма
normalizeDF <- function(d){
  maxspeed <- max(d[,3]); minspeed <- min(d[,3])
  maxram <- max(d[,5]); minram <- min(d[,5])
  maxhd <- max(d[,4]); minhd <- min(d[,4])

  newspeed <- (d$speed - minspeed)*100/(maxspeed - minspeed)
  newram <- (d$ram - minram)*100/(maxram - minram)
  newhd <- (d$hd - minhd)*100/(maxhd - minhd)

  normDF <- data.frame(newspeed, newram, newhd)
  rownames(normDF) <- rownames(d)
  colnames(normDF) <- c("speed", "ram", "hd")
  return(normDF)
}
```

```r
# Функция нормировки идеальной точки
normalizePoint <- function(p, d){
   maxspeed <- max(d[,3]); minspeed <- min(d[,3])
  maxram <- max(d[,5]); minram <- min(d[,5])
  maxhd <- max(d[,4]); minhd <- min(d[,4])

  newspeed <- (p[1] - minspeed)*100/(maxspeed - minspeed)
  newram <- (p[2] - minram)*100/(maxram - minram)
  newhd <- (p[3] - minhd)*100/(maxhd - minhd)

  normPoint <- c(newspeed, newram, newhd)
  return(normPoint)
}
```

```r
DPareto <- function(X, Y){
  p <- TRUE; l <- FALSE; i <- 1;
  while(p & (i <= length(X))){
    if(X[i] < Y[i]) p <- FALSE
    if(X[i] > Y[i]) l <- TRUE
    i <- i+1
  }
  if(!p | !l) return (FALSE)
  else return (TRUE)
}
```

```r
# Функция евклидого расстояния
distance <- function(A, B){ return(sqrt(sum((A-B)^2))) }
```
